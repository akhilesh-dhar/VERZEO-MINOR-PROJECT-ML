# 4)Which year had the minimum sales for "Nike" products?

```python
df['order year'] = pd.DatetimeIndex(df['order date (DateOrders)']).year
```

```python
df['order year'].value_counts()
```

```python
df['Date'] = pd.to_datetime(df['order date (DateOrders)'])
df.groupby(df['Date'].dt.year).sum()
```

```python
#ADDING A FEATURE WHICH SHOWS WHERE "NIKE " IS PRESENT
df["Index"]= df["Product Name"].str.find('Nike')
df.head(51)
```

Calculating the sales of "nike" products for the year 2015, 2016, 2017 and 2018

```python
D=(df['Sales'].where(df['Index']== 0).where(df['order year']==2015))
D=pd.DataFrame(D)
D.dropna()
D.sum()
```

```python
E=(df['Sales'].where(df['Index']== 0).where(df['order year']==2016))
E=pd.DataFrame(E)
E.dropna()
E.sum()
```

```python
F=(df['Sales'].where(df['Index']== 0).where(df['order year']==2017))
F=pd.DataFrame(F)
F.dropna()
F.sum()
```

```python
G=(df['Sales'].where(df['Index']== 0).where(df['order year']==2018))
G=pd.DataFrame(G)
G.dropna()
G.sum()
```

## ANSWER: MIN SALE WAS MADE IN THE YEAR 2018
